### Shifts taken by player Patrick

```python
def totalShifts(player_id):
    playerShifts = shiftsOfBothPlayers.loc[shiftsOfBothPlayers.player_id == player_id]
    playertotalShiftPerGame = pd.DataFrame({'Total_Shifts' : playerShifts.groupby( [ 'game_id'] ).size()}).reset_index()
    return playertotalShiftPerGame
```

```python
print("Total number of shifts taken by Patrick per game are:")
patrickShifts = totalShifts(8474141)
patrickShifts.head()
```

```python
print("Total games Patrick played for season 2016-2017 is",patrickShifts.shape[0] )
```

### Powerplay percentage calculation (Power Play Attempts / Power Play Goals)

```python
teamPlayedGames['powerPlayPercentage'] = round((teamPlayedGames.powerPlayGoals/teamPlayedGames.powerPlayOpportunities)*100, 2)
teamPlayedGames = teamPlayedGames.fillna(0.00)   
```

```python
teamPlayedGames['powerPlayPercentage']= teamPlayedGames['powerPlayPercentage'].astype(str)+'%'
```

```python
teamPlayedGames['turnOverDifferential'] = teamPlayedGames.takeaways - teamPlayedGames.giveaways
```

```python
teamPlayedGames.head()
```

### Visualize the shots and hits of the team

```python
teamPlayedGames.plot(x="game_id", y=["shots", "hits"], kind="bar")
```
